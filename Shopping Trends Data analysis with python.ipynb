```python
data=pd.read_csv(r'C:\Users\User\Desktop\data_hub\shopping_trends.csv')
```

```python
data.head()
```

```python
data.sort_values('Review Rating',ascending=False,inplace=True)
```

```python
data.head()
```

```python
Category_Review_Rating=data.groupby('Category',as_index=False)['Review Rating'].mean().sort_values(by='Review Rating',ascending=False)
Category_Review_Rating
```

```python
ax=sns.barplot(x='Category',y='Review Rating',data=Category_Review_Rating,palette='RdYlGn')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Average Review Rating by Category')
plt.show()
```

```python
data['Subscription Status'].value_counts().plot(kind='pie',autopct='%.1f%%')
plt.title('Pie Chart of Subscription')
plt.ylabel('')
plt.show()
```

```python
#Color that sales maximum
plt.figure(figsize=(18,4))
sns.countplot(x='Color',data=data,palette='Set2')
plt.title('Color that sales maximum')
plt.xticks(rotation=70)
plt.show()
```

```python
#showing color that sales 
data['Color'].value_counts()
```

```python
#Scatter Plot - Previous Purchases Review Rating
data.head(1000).plot(x='Previous Purchases',y='Review Rating',kind='scatter')
plt.title('Top 1000 items Scatter Plot - Previous Purchases Review Rating')
plt.show()
```

```python
Season_Purchase_Amount=data.groupby('Season',as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False)
Season_Purchase_Amount
```

```python
sns.barplot(x='Season',y='Purchase Amount (USD)',data=Season_Purchase_Amount,palette='Spectral')
plt.title('Total sales in season wise')
plt.show()
```

```python
plt.figure(figsize=(16,4))
data.boxplot(column='Purchase Amount (USD)',by='Frequency of Purchases')
plt.title('Box plot of Purchase Amount based on Frequency of Purchases')
plt.xticks(rotation=70)
plt.show()
```

```python
data['Preferred Payment Method'].value_counts()
```

```python
#data visualization in Preferred Payment Method
ax=sns.countplot(x='Preferred Payment Method',data=data,palette='YlGnBu')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Most Preferred Payment Method in Purchasing')
plt.xticks(rotation=70)
plt.show()
```

```python
#Payment Method in Purchase Amount (USD) in tatal
data.groupby('Payment Method')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```
